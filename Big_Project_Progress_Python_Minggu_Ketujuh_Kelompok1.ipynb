```python
max_features = 1000
databaru= data['Tweet_stem'].astype(str)


tfidf = TfidfVectorizer(max_features=max_features, ngram_range=(1,3), smooth_idf=False)
tfs = tfidf.fit_transform(databaru)
IDF_vector = tfidf.idf_

# hitung TF x IDF sehingga dihasilkan TFIDF matrix / vector
tfidf_mat = tfidf.fit_transform(databaru).toarray()
```

```python
X1=vec_TF_IDF.transform(databaru).toarray()
data_tfidf=pd.DataFrame(X1,columns=vec_TF_IDF.get_feature_names_out())
data_tfidf.head(5)
```

```python
from textblob import TextBlob

def getSubjectivity(review):
    return TextBlob(review).sentiment.subjectivity

def getPolarity(review):
    return TextBlob(review).sentiment.polarity

def analyze(score):
    if score < 0:
        return '1'
    else:
        return '0'

data_fix = pd.DataFrame(data[['casefold']])
data_fix['Subjectivity'] = data_fix['casefold'].apply(getSubjectivity)
data_fix['Polarity']=data_fix['casefold'].apply(getPolarity)
data_fix['Target']=data_fix['Polarity'].apply(analyze)
data_fix
```

```python
data_fix.info()
```

```python
data_fix["Target"] = data_fix["Target"].astype(str).astype(int)
```

```python
Y=data_fix['Target'].values
```

```python
# Feature selection
x_train=np.array(data_tfidf)
y_train=np.array(Y)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

Feature_Chi2=SelectKBest(chi2, k=2500)
x_kbest_feature=Feature_Chi2.fit_transform(x_train, y_train)

# Reduced feature
print('Original feature number', x_train.shape[1])
print('Reduced feature number', x_kbest_feature.shape[1])
```

```python
data3=pd.DataFrame(Feature_Chi2.scores_, columns=["Nilai"])
data3.head(5)
```

```python
# Menampilkan feature beserta nilainya
feature=vec_TF_IDF.get_feature_names_out()
feature

data3['Fitur']=feature
data3.head(5)
```

```python
# mengurutkan nilai fitur terbaik
data3.sort_values(by='Nilai', ascending=False)
```

```python
mask=Feature_Chi2.get_support()
mask
```

```python
#menampilkan fitur yang terpilih berdasarkan nilai mask atau nilai tertinggi yang sudah ditetapkan pada chi square

fitur_baru = []
for bool, f in zip(mask, feature):
  if bool :
    fitur_baru.append(f)
  selected_feature = fitur_baru
selected_feature
```

```python
#membuat vocabulary baru berdasarkan fitur yang terseleksi

fitur_baru_seleksi = {}

for (k,v) in vec_TF_IDF.vocabulary_.items():
  if k in selected_feature:
    fitur_baru_seleksi[k]=v

fitur_baru_seleksi
```

```python
x2 = selected_x
y2 = Y
```

```python
#mendefinisikan data train sebesar 80% dan data test sebesar 20% untuk membentuk model ANN
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x2, y2, test_size = 0.2, random_state = 1)
```

```python
import tensorflow as tf
import numpy as np
```

```python
ANN = tf.keras.models.Sequential()
ANN.add(tf.keras.layers.Dense(units=6,activation="relu"))
ANN.add(tf.keras.layers.Dense(units=6,activation="relu"))
ANN.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
ANN.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
ANN.fit(X_train,Y_train,batch_size=32,epochs = 100)
```

```python
evaluation = ANN.evaluate(X_test, Y_test)
```
